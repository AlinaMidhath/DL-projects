**Converting an RGB img to a grayscale img that is a black and white img using Opencv**


opencv is an important library alot of image recognition and computer vision task can be done using opencv

It is easy to work on a grayscale img when compared to a RGB img

The maain reason for converting this is the RGB image requires 3 matrices whereas the graysale image requires only one matrix , so the conversion will save alot of memory in case of big datasets

```python
#importing opencv  library
import cv2
```

```python
img = cv2.imread('/content/dog.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_img )
```

```python
grayscale_img.shape
```

Now the shape is not showing 3 bcs our img array has been converted from 3 channel(RGB) to 1 channel(grayscale)


```python
#displaying the img using numpy array
img_plot = plt.imshow(grayscale_img)
plt.show()
```
